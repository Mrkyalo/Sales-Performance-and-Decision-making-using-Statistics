# Central Limit Theorem
- Draw 200 samples of size n = 30
- Plot distribution of sample means
- Explain why it becomes normal

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('CLT.csv')
df.info()
```

```python
print(df.columns)
```

```python
revenue = df['revenue'].dropna()
```

```python
revenue=pd.to_numeric(revenue, errors='coerce')
revenue = revenue.dropna()
```

```python
#Parameters
num_samples=200
sample_size=30
```

```python
#store means of samples
sample_means=[] 
```

```python
# Generate sample data
for i in range(num_samples):
    sample = revenue.sample(sample_size, replace=True)
    mean = sample.mean()
    sample_means.append(mean)
```

```python
#Plotting the distribution of sample means
plt.hist(sample_means, bins=15, edgecolor='black')
plt.title('Distribution of Sample Means (n=30)')
plt.xlabel('Sample Mean Revenue')
plt.ylabel('Frequency')
plt.show()
```

#### Explain why the CLT becomes normal
- The distribution of sample means approaches normal distribution as the sample size gets larger regardless of the population distribution

#### Task 5.1 – Hypotheses
H₀: Mean revenue (campaign) = Mean revenue (no campaign)
H₁: Mean revenue (campaign) > Mean revenue (no campaign)
* Specify:

One-tail test
Confidence level = 95%
Alpha = 0.05
Reference: Hypothesis formulation slides

- Null Hypothesis(h0) suggests there is no difference in mean revenue if the company undertakes marketing campaign or no marketing campaign
- Alternative hypothesis (h1) suggests that the mean revenue after marketing campaign is greater than mean revenue if there is no marketiing campaign

- One-tail test : We are only interested in marketing campaign revenue being greater than no campaign revenue
- Confidence level =95% : We are 95% confident in our conclusion that marketing campaign revenue is greater than no campaign revenue.There is only 5% chance of making a wrong rejection
- Alpha = 0.05 -This the probability of rejecting the null hypoyhesis when it is true.It represents the risk of making a Type 1 error(false positive)
- Conclusion: We reject the null hypothesis.There is sufficient eveidence at 95% confidence level that the mean revenue from campaign is greater than the mean revenue without the campaign.

Task 6.1 – Errors
Explain in context:
Type I Error
Type II Error
Example:
What would a Type I error mean for the company?
Reference: Type I vs Type II slides

- Type I Error- Also known as 'False positive' occurs in hypothesis testing when the null hypothesis(H0) is true but is incorrectly rejected.
    - Examlpe: A medical test indicates a patient has a disease when they do not.
    - Causes: Random sampling error or setting the significance level too high

- Type II Error- Also 'False negative' occurs in hypothesis testing when null hypothsis is not rejected even though it is false.
    - Example: Failing to diagonise a disease when it is actually present
    - Causes: Insufficient sample sizes,high data variability

- What would a Type I error mean for the company?
    - Type I error for the company would mean the company will believe there is no change in mean revenue if it undertakes marketing campaign or not.This will cause company not meeting its set target to make revenue for the year.
